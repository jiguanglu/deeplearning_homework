## 我們先取前 50 筆資料，作為實驗素材

```python
import numpy as np
import pandas as pd
pd = pd.read_csv("./pokemon.csv")
newcp = np.array(pd['cp_new'])
cp = np.array(pd['cp'])
cp = cp[:50]
newcp = newcp[:50]

import matplotlib.pyplot as plt
# plt.plot(newcp,cp)
plt.scatter(cp, newcp)
# plt.ylabel('some numbers')
plt.show()
```

```python
import keras
```

```python
from keras.layers import Dense
```

```python
from keras.layers import Input, Dense
from keras.models import Model
from keras import regularizers
from keras import initializers
from keras import optimizers


# This returns a tensor
x = Input(shape=(1,))
y = Dense(1, 
          kernel_initializer=keras.initializers.Ones(), 
          activation='linear',
          kernel_regularizer=regularizers.l1(2),)(x)
```

```python
model = Model(inputs=x, outputs=y)
model.compile(optimizer=optimizers.SGD(lr=0.000001),
              loss='mean_squared_error',)
```

```python
model.summary()
```

```python
for _ in range(1000):
#     for now_cp,now_newcp in zip(cp,newcp):
        
#         now_cp = np.array([now_cp]).reshape([1,])
#         now_newcp = np.array([now_newcp])
#         model.train_on_batch(now_cp,now_newcp)
    model.fit(cp,newcp)
#     print(model.evaluate(cp,newcp))
```

```python
predict_cp = model.predict(cp)
```

```python
plt.scatter( cp,predict_cp)
plt.scatter( cp,newcp)

# plt.ylabel('some numbers')
plt.show()
```
